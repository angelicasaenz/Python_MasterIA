### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio
suma = 0
for i in esperanzas_de_vida:
    suma += i

promedio = suma / len(esperanzas_de_vida)
print("El promerido es: ", promedio)

# Máximo

maximo = esperanzas_de_vida[0]
for i in esperanzas_de_vida:
    if i > maximo:
        maximo = i

print("El maximo es: ", maximo)
# Mínimo

minimo = esperanzas_de_vida[0]
for i in esperanzas_de_vida:
    if i < minimo:
        minimo = i
print("El minimo es: ", minimo)

````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí

def clasificar_esperanza_vida(num):
    if num >= 75:
        return "Alta"
    elif num >= 60 and num <= 74.9:
        return "Media"
    else:
        return "Baja"
    
print(clasificar_esperanza_vida(75))
print(clasificar_esperanza_vida(74.5))
print(clasificar_esperanza_vida(8))
print(clasificar_esperanza_vida(60))
print(clasificar_esperanza_vida(80))
````

### 2.4 — Filtrado y selección

🔧 **Ejercicio 2.4**
1. Filtra los países de **América** (`region == "Americas"`) y muestra solo las columnas: `pais`, `esperanza_vida`, `gasto_salud_usd`
2. Filtra los países con esperanza de vida **mayor a 80 años** — ¿cuántos son? ¿De qué regiones son?
3. Filtra los países con **nivel_ingresos == "Low"** y calcula su promedio de `mortalidad_infantil`

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí

# 1. Esperanza de vida promedio por region

vida_region = salud_mundial.groupby("region")["esperanza_vida"].mean()
print(f"Esperanza de vida promedio por región:\n{vida_region}")

# 2. Gasto en salud promedio por nivel_ingresos

gasto_ingresos = salud_mundial.groupby("nivel_ingresos")["gasto_salud_usd"].mean()
print(f"\nGasto en salud promedio por nivel de ingresos:\n{gasto_ingresos}")

# 3. Número de países por región
paises_region = salud_mundial.groupby("region")["pais"].count()
print(f"\nNúmero de países por región:\n{paises_region}")

# 4. Región con mayor y región  con menor esperanza de vida promedio
mayor_region = vida_region.idxmax()
menor_region = vida_region.idxmin()
print(f"\nRegión con mayor esperanza de vida promedio: {mayor_region}")
print(f"Región con menor esperanza de vida promedio: {menor_region}")
````
